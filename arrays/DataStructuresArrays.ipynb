# Slicing in an array
is a process of getting a sequence from array from a strating index to ending index

```python
x = list(range(0,100,2))
arr = array('i' , x)

print(arr)
print('\n')

#slicing operation
#get first 20 elements
arr0 = arr[:20]
print(arr0)
print(len(arr0))
print("\n")

#get last 20 elements
arr1 = arr[-20:]
print(arr1)
print(len(arr1))
print("\n")


#negative indexing
arr2 = arr[10:-10]
print(arr2)
print(len(arr2))
print("\n")

#reversing the order using slicing
arr3 = arr[::-1]
print(arr3)
print(len(arr3))
print("\n")


```

```python
def search(thelist , n):
    for i in range(0, len(thelist)):
        if(thelist[i] == n):
            return True

z = list(range(0,1000,3))



x = int(input("Enter number you want to search: "))

if(search(z , x)):
    print("item found at: ", z.index(x))
else:
    print("item not found ")
```

## Binary Search
Note:
    in binary search all the values have to be sorted
    

```python
def binarySearch(theList, n):
    lowerBound = 0
    upperBound = len(theList) - 1
    
    while lowerBound <= upperBound:
        mid = (upperBound + lowerBound)//2
        mid_num = theList[mid]
        
        if mid_num == n:
            return mid
        if mid_num < n:
            lowerBound = mid + 1
        else:
            upperBound = mid - 1
    return -1
    
    
    
    
    

z = list(range(0,10000,3))



x = int(input("Enter number you want to search: "))

index = binarySearch(z , x)

if(index != -1):
    print("item found at: ", index)
else:
    print("item not found ")
```

```python
import time

def time_func(func):
    def wrapper(*args, **kwargs):
        start= time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(func.__name__ + " Took " + str((end-start)*1000) + " Milliseconds")
        return result
    return wrapper




#Decorating the Binary Search function
@time_func
def binarySearch(theList, n):
    lowerBound = 0
    upperBound = len(theList) - 1
    
    while lowerBound <= upperBound:
        mid = (upperBound + lowerBound)//2
        mid_num = theList[mid]
        
        if mid_num == n:
            return mid
        if mid_num < n:
            lowerBound = mid + 1
        else:
            upperBound = mid - 1
    return -1
    
        
#Decorating the Linear Search Function
@time_func    
def linear_search(num_list, item):
    for index, element in enumerate(num_list):
        if element == item:
            return index
    return -1    



z = list(range(0,10000001,3))

x = int(input("Enter number you want to search: "))

index = linear_search(z , x)


index = binarySearch(z , x)
```

## Can Use Recursion for Binary Search


```python
def binarySearchRecursive (num_list, num_Find, lowerBound, upperBound):
    if (upperBound < lowerBound):
        return -1
    
    mid = (upperBound + lowerBound)//2
    mid_num = num_list[mid]

    if mid_num == num_Find:
        return mid
    
    if mid_num < num_Find:
        lowerBound = mid + 1
    else:
        upperBound = mid - 1
        
    return binarySearchRecursive(num_list, num_Find, lowerBound, upperBound)
  



z = list(range(0,100000,3))

x = int(input("Enter number you want to search: "))

index = binarySearchRecursive(z , x, 0, len(z)-1)

if(index != -1):
    print("item found at: ", index)
else:
    print("item not found ")
        
```

# BUBBLE SORT

```python
@time_func
def bubbleSort(elements):
    size = len(elements)
    i=j=0
    for j in range(size-1):
        for i in range(size-1-j):
            if elements[i] > elements[i+1]:
               
                temp = elements[i]
                elements[i] = elements[i+1]
                elements[i+1] = temp
    return elements

arr = array("i", [3,54,678,234,76,234,568,45,1213,76,87,34,80])

print(arr, "\n")

sorted_array = bubbleSort(arr)

print(sorted_array)



```

# Quick Sort

```python
def swap(a , b , arr):
    if a != b:
        tmp = arr[a]
        arr[a] = arr[b]
        arr[b] = tmp


def partition(items, start, end):
    pivot_index = start
    pivot = items[pivot_index]
      
    while start < end:
        while items[start] <= pivot:
            start += 1

        while items[end] > pivot:
            end -= 1

        if start < end:
            swap(start, end, items)
    
    swap(pivot_index, end, items)
    return end
    
def quickSort(items, start, end):
    if start<end:
        p1 = partition(items,start,end)
        quickSort(items, start, p1-1) #left partition
        quickSort(items, p1+1, end) #right partition
    
    
    
x=[11,23,456,1,2,4,67,9,0]
print(x)
quickSort(x, 0 , len(x)-1)
print(x)

```

# Merge Sort

```python
def mergeSort(arr):
    if len(arr) <=1:
        return arr
    
    mid = len(arr)//2
    
    left = arr[:mid]
    right = arr[mid:]
    
    left = mergeSort(left)
    right = mergeSort(right)
    
    return merge_sorted_arrays(left,right)
    
    
#code to merge two sorted array

def merge_sorted_arrays(a,b):
    sortedList = []
    len_a = len(a)
    len_b = len(b)
    i=j=0
    
    while i < len_a and j < len_b:
        if a[i] <= b[j]:
            sortedList.append(a[i])
            i += 1
        else:
            sortedList.append(b[j])
            j += 1
    while i< len_a:
        sortedList.append(a[i])
        i += 1
    while j < len_b:
        sortedList.append(b[j])
        j+=1
    
    return sortedList


x=[11,23,456,1,2,4,67,9,0]

print(mergeSort(x))

```

# Insertion sort

```python
def insertionSort(items):
    for i in range (1, len(items)):
        anchor = items[i]
        j = i - 1
        while j >= 0 and anchor < items[j]:
            items[j+1] = items[j]
            j = j-1
        items[j+1] = anchor
    
    
x=[11,23,456,1,2,4,67,9,0,100,99,89]
print(x)
insertionSort(x)
print(x)
```

# Shell Sort
optimization of insertion sort

```python
def shellSort(items):
    size = len(items)
    gap = size//2
    
    while gap > 0:
        for i in range(gap , size):
            anchor = items[i]
            j = i
            while j>=gap and items[j-gap] > anchor:
                items[j] = items[j-gap]
                j -= gap
            items[j] = anchor    
        gap = gap // 2
   

z = [11,23,456,1,2,4,67,9,0]

tests = [
    [9,78,61,53,23,21,17,12,9,7,6,2,1],
    [],
    [1,5,8,9],
    [234,3,1,56,34,12,9,12,1300],
    [5]
    
]


shellSort(z)

print(z)



for each in tests:
    shellSort(each)
    print(each)

```

# Selection Sort

```python


def selectionSort(items):
    size = len(items)
    for i in range(size):
        minIndex = i
        for j in range(minIndex + 1, size):
            if items[j] < items[minIndex]:
                minIndex = j
       
        if i != minIndex:
            items[i] , items[minIndex] = items[minIndex], items[i]



z = [11,23,456,1,2,4,67,9,10]

selectionSort(z)

print(z)

tests = [
    [9,78,61,53,23,21,17,12,9,7,6,2,1],
    [],
    [1,5,8,9],
    [234,3,1,56,34,12,9,12,1300],
    [5]
    
]



for each in tests:
    shellSort(each)
    print(each)
```

# 1 Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target

You may assume that each input would have exactly one solution and may not use the same element twice

You can return your answer in any order

Example 1

Input: num = [2,7,11,15], target = 9
output: [0,1]
    
Example 2
input: num = [3,2,4], target = 6
output: [1,2]
    
example 3
input: num = [3,3], target = 6
output: [0,1]


```python


def sum_pairs(A, target, k) -> list:
    if k > len(A):
        return -1

    current_sum = 0
    for i in range(0, len(A)-1):
        
        current_sum += A[i]
        if i >= k - 1:
            if current_sum == target:
                return [i, i - k + 1]
            else:
                current_sum -= A[i - (k - 1)]
    return sum_pairs(A, target, k + 1)


num = [2, 7, 11, 15]
target = 26
win_size = 2

print(sum_pairs(num, target, win_size))
```

ith customer has in the jth bank. Return the wealth that the richest customer has.

A customer's wealth is the amount of money they have in all their bank accounts. 
The richest customer is the customer that has the maximum wealth.

Input: accounts = [[1,2,3],[3,2,1]]

Both customers are considered the richest with a wealth of 6 each, so return 6.

Explanation: 
1st customer has wealth = 6
