# STACKS 

## -- Stacks are abstract data typers
 -- They are implemented using other data types Eg arrays, linked_lists
 -- 
 -- Stacks allows access to only oe data item: the last item inserted
 -- To access the second item you need to remove the last item 
 -- As a result it is a Last in First out order
    
 -- it is usefull in many situations such as:
    - Compiler program which wants to check if the parentheses, braces and brackets
       are balanced in source code i.e the last bracket should be closed first
    
    - In arithemetic calculations eg 3*(3+4) where last set of brackets are computed first
    
    - In microprocessor where a set of nested function calls are inserted in a stack and 
      the last one is executed first
        
    - Implement undo/redo functionality
    
        
SOME TERMINOLOGIES

-Pushing: placing item on top of the stack
-Popping: Removing item from top of the stack
-Peeking: Vieiwng the top of the stack


# Stack implementation using Lists

```python
class Stack:
    def __init__(self):
        self.stack = []
        
    def pop(self):
        if self.is_empty():
            return
        return self.stack.pop()
    
    def push(self, data):
        self.stack.append(data)
    
    def is_empty(self):
        if len(self.stack) == 0:
            return True
        return False
        
    def peek(self):
        if self.is_empty():
            print("Stack is empty ")
        else:
            print(self.stack[-1])
    
    def print_stack(self):
        print_str = ' '
        if self.is_empty() is False:
            for each in self.stack:
                print_str += str(each) + " ----- "
            print(print_str)
        else:
            print("Stack is Empty ")
            
 
    
    
```

```python

newStack = Stack()

while True:

    print("MENU")
    print("1 to push items ")
    
    print("2 to peek items ")
    print("3 to pop items ")
    print("4 to print all items ")
    print("5 to check if stack is empty ")
    print("10 to quit ")

    option = int(input("Enter your choice of action "))
    print()
    if option == 1:
        item = input("Enter the item to push into stack ")
        newStack.push(item)
    elif option == 2:
        newStack.peek()
    elif option == 3:
        newStack.pop()
    elif option == 4:
        newStack.print_stack()
    elif option == 5:
        if(newStack.is_empty()):
            print("Stack is empty")
        else:
            print("Stack not empty ")
    elif option == 10:
        break
    else:
        print("wrong input ")
    print()
    
            
    
```

# LinkedList Implementation

```python
class Node:
    def __init__ (self, data):
        self.data = data
        self.next = None
        

##############################################################################
#
#  Initialize the stack and put set head to none 
#
##############################################################################
class Stack:
    def __init__(self):
        self.head = None
        
        
##############################################################################
#
#  pushing involves inserting item in the stack
#   Requires the data as an argument
#
##############################################################################
    def push(self, data):
        if self.head is None:
            self.head = Node(data)
        else:
            new_node = Node(data)
            new_node.next = self.head
            self.head = new_node 
    
##############################################################################
#
#  Pop refers to removing the last item that was inserted
#     since we keep the items at the head 
#       we just see the item at the head and point the head to its next item
#
#
##############################################################################

    def pop(self):
        if self.head is None:
            print("stack empty ")
        else:
            popped = self.head.data
            self.head = self.head.next
            return popped
    
##############################################################################
#
#   Peek refers to seeing the item which was last inserted in the stack
#    We can just do that by printing the item at the head
#
##############################################################################
    def peek(self):
        if self.head is None:
            print("Stack empty ")
        else:
            print(self.head.data)
```
