# Linked Lists

(linear data sturctures)

Sequnce of data structure, which are connected together via links
They contain item, and link for the next list
Second most used data structure after array

They are not stored at a contigious location. The elements are linked using pointers





# Why use Linked Lists and not arrays

-The size of array is fixed. in arrays the elements must be stored contiguously
-Inserting a new element in array is costly, room has to be created. if sorted, shifting of
older elements is required
-Deletion is expensive also

DrawBacks of LinkedList
-Random access is not efficient
-Extra memory space to store pointers
-Not cache friendly because its not saved contiguously


# Some terms

Nodes: A container for some data
Pointer: Reference to another node
Head: The very first of node of the linked list
Tial: The very last node of the linked list

# Doubly Linked List

# Advantages over single linked list

-Traversed and searched in both direction
-Basic operation like insertion and deletion are earsier
because we dont need to traverse to the predecessor node

disadvantages
-More memory space to store one extra reference
-few additional steps are required to be performed for 
insertion and deletion operation


```python
#First we code the Node class

class Node:
    def __init__ (self ,data):
        self.item = data
        self.next = None
        self.prev = None
```

```python
#Next we create DoublyLinkedList class for all 
# DLL related function
```

```python
class DoublylinkedList:
    def __init__ (self):
        self.head = None
        
########################################################
#
# Inserting the element in an empty list
#   -Check if the list is empty
#   -if empty: node is initialized,
#     the self.head is set the node created
#
########################################################

    def insert_in_empty_list (self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
        else:
            print("List is not empty ")
        
########################################################
#
# Inserting the element at the start
#     - First check whether list is empyt or not
#     - if the list is empty:
#        -- use logic for insert_in_empty
#     - else three operations are performed
#     1. for new node: reference to next node will be set to current head
#     2: for head the prev node will be set to newly create node
#     3: finally current head will be set to new node
#
########################################################

    def insert_at_start(self, data):
        if self.head is None:
            
            #
            # insert_in_empty_list(data)
            #
            new_node = Node(data)
            self.head = new_node
            print("Node inserted ")
            return
        new_node = Node(data)
        new_node.next = self.head
        self.head.prev = new_node
        self.head = new_node
        
##########################################################
#
# Inserting the element in at end
#  -if list is empty logic == insert_in_emptyList
#  -else traverse till node with next is Null
#   then
#     -- prev ref for new node is set to last node
#     -- the next ref of last node is set to new node
#
########################################################

    def insert_at_end(self, data):
        if self.head is None:
            new_node = Node(data)
            self.head = new_node
            return
        
        pointer = self.head
        while pointer.next is not None:
            pointer = pointer.next
        new_node = Node(data)
        pointer.next = new_node
        new_node.prev = pointer

########################################################
#
# Inserting the element after another element
#  -If list is empty simply say list is empty
#  -iterate to get the node 
#    Then perform 4 steps
#      --set prev of new node to selected node
#      --set next of new node to next of selected node
#      if the selected node is not last
#         --set prev of next node from selected node to new node
#      finally set the next of selected node to NEW node
#
########################################################

    def insert_after_item(self, element, data):
        if self.head is None:
            print("List is empty ")
            return
        else:
            pointer = self.head
            while pointer is not None:
                if pointer.item == element:
                    break
                pointer = pointer.next
                
            if pointer is None:
                print("Item not in list ")
            else:
                new_node = Node(data)
                new_node.prev = pointer
                new_node.next = pointer.next
                if pointer.next is not None:
                    pointer.next.prev = new_node
                pointer.next = new_node
                

                
########################################################
#
# Inserting the element before another item
#   -if list empty display list is empty
#   -iterate to find selected node
#   -if found perform 4 steps
#     -- set next of new node to selected node
#     -- set prev of new node to prev of selected node
#     -- if selected node's prev is not null     
#       -- set next of node prev to selected to new node
#     -- set prev of selected to new node
#
########################################################

    def insert_before_item(self, element, data):
        if self.head is None:
            print("List is empty ")
            return
        else:
            pointer = self.head
            while pointer is not None:
                if pointer.item == element:
                    break
                pointer = pointer.next
            if pointer is None:
                print("Element not found ")
            else:
                new_node = Node(data)
                new_node.next = pointer
                new_node.prev = pointer.prev
                if pointer.prev is not None:
                    pointer.prev.next = new_node
                pointer.prev = new_node
                
########################################################
#
# Travesing the DLL
#   Similar to travesing single linked list
#
#
########################################################
                    

    def traverse_list(self):
        if self.head is None:
            print("List has no Elements ")
            return
        else:
            pointer = self.head
            while pointer is not None:
                print(pointer.item, " ")
                pointer = pointer.next
                
########################################################
#
#  Deleting from start
#  - easiest
#   - Perform 2 checks:
#      --see if list is empty : nothing to delet
#      -- if list has only one element:
#          --- simply set head to none
#    -- Set value of head to next node
#    -- Set prev of head to null
#    
#
########################################################

    def delete_at_start(self):
        if self.head is None:
            print("The list is empty ")
            return
        if self.head.next is None:
            self.head = None
            return
        self.head = self.head.next
        self.head.prev = None
        
########################################################
#
# Delete from End
#  -- Check if list is empty
#  -- Check if list has only one element
#     -- set head to None
#  -- if list has many elements iterate to end
#     -- set next of prev node to last == null
#
########################################################

    def delete_at_end(self):
        if self.head is None:
            print("No element to delete ")
            return
        if self.head.next is None:
            self.head = None
            return
        pointer = self.head
        while pointer.next is not None:
            pointer = pointer.next
        pointer.prev.next = None
        
########################################################
#
# Deleting by element
#  -Trickiest of the all deletion function
#   - check if list empty
#   -if list has single node and is the one to delete
#      --- self head to Null
#   - if list has more node and first one is to delete
#       ----Same as delete at begining
#   -traverse to find element to delete then:
#     --Set value of next of prev node to next of node to delete
#     --set prev of next node to prev of node to delete
#   if node to delete is at end
#      --set next of prev to last node to null
#
########################################################

    def delete_element_by_value(self, element):
        if self.head is None:
            print("The list has no element to delete")
            return 
        if self.head.next is None:
            if self.head.item == element:
                self.head = None
            else:
                print("Item not found")
            return 

        if self.head.item == element:
            #You can call the function for deleting at the begining
            self.head = self.head.next
            self.head.prev = None
            return

        n = self.head
        while n.next is not None:
            if n.item == element:
                break;
            n = n.next
        if n.next is not None:
            n.prev.next = n.next
            n.next.prev = n.prev
        else:
            if n.item == element:
                n.prev.next = None
            else:
                print("Element not found")
                
########################################################
#
# Reversing a List
#  -Perform following operation
#    -- next of start node should be none coz head will be last node 
#    -- Prev of last node should be none 
#    -- the nexts of nodes execpt(head & tail) be swapped with prevs 
#
########################################################

    def reverse_linked_list(self):
        if self.head is None:
            print("List has no elements")
            return
        pointer = self.head
        second_pointer = pointer.next
        pointer.next = None
        pointer.prev = second_pointer
        while second_pointer is not None:
            second_pointer.prev = second_pointer.next
            second_pointer.next = pointer
            pointer = second_pointer
            second_pointer = second_pointer.prev
        self.head = pointer
```

# CIRCULAR LINKED LIST

```python
class Node:
    def __init__ (self, data):
        self.data = data
        self.next = None
        
    

class CircularLinkedList:
    def __init__ (self):
        self.head = None
        
##############################################################
#
#    function below is to check whether list is empty of not
#
##############################################################

    def empty(self):
        if self.head is None:
            return True
        else:
            return False

############################################################################
#
#    Traversing and displaying element in list
#    -- since the tail node's next pointer is the head node:
#     -- we loop and print each node till our pointer reaches the head node
#
############################################################################

    def DisplayList(self):
        pointer = self.head
        if(pointer != None):
            print()
            print("Circular linked list: ", end=" ")
            print()
            while(True):
                print(pointer.data, end=" ")
                pointer = pointer.next
                if(pointer == self.head):
                    break
            print()
        else:
            print("The List is empty ")
        
        
        
##############################################################
#
#    Adding data into the Circular linked list in the begining
#     -- initially store address of current 1st node to new node
#     --point last node to new node
#
##############################################################
    
    
    def add_begining(self, data):
        new_node = Node(data)
        if self.empty() is True: 
            self.head = new_node
            new_node.next = self.head
            return
        else:
            pointer = self.head
            while (pointer.next != self.head):
                pointer = pointer.next
            pointer.next = new_node
            new_node.next = self.head
            self.head = new_node
    
    
    
##############################################################
#
#    Adding data into the Circular linked list at the end
#     --Store address of head node to next of new node 
#     --point current last node to new node
#     --Make the new node the last node
#
##############################################################
            
            
            
    def insert_at_end(self, data):
        new_node = Node(data)
        if(self.head == None):
            self.head = new_node
            new_node.next = self.head
            return
        else:
            pointer = self.head
            while(pointer.next != self.head):
                pointer = pointer.next
            pointer.next = new_node
            new_node.next = self.head
        

##############################################################
#
#    Adding data into the Circular linked list at the given position
#     --Go to the node after which new node has to be inserted
#     --Point next of new node to the next node after which new node has to be inserted
#     --Store address of new node at the next of the node after which new node has to be inserted
#
##############################################################
          
    def Insert_at_position(self, element, position):     
        newnode = Node(element)
        pointer = self.head
        count = 0
        if(pointer != None):
            count += 1
            pointer = pointer.next
            
        while(pointer != self.head):
            count += 1
            pointer = pointer.next
    
        if(position < 1 or position > (count+1)):
            print("\nInavalid pos.")
        elif (position == 1):
            if(self.head == None):
                self.head = newnode
                self.head.next = self.head
            else:
                while(pointer.next != self.head):
                    pointer = pointer.next
                newnode.next = self.head
                self.head = newnode
                pointer.next = self.head
        else:
            pointer = self.head
            for i in range(1, position-1):
                pointer = pointer.next 
            newnode.next = pointer.next
            pointer.next = newnode
            


###############################################################
#
#    Deleting data in begining in circular list
#     --if head is not null and List only has only element
#       -- set the current head to Null
#     -- if list has more than one element
#      -- set pointer to head and another variable for head
#       --loop pointer as long as pointer is not head
#       set next of head to the current head.next
#      then set the pointer.next to current head
#      Clear the node1 var
#
##############################################################
 
    def Del_at_Begining(self):
        if(self.head != None):
            if(self.head.next == self.head):
                self.head = None
            else:
                pointer = self.head
                node1 = self.head
            while(pointer.next != self.head):
                pointer = pointer.next
            self.head = self.head.next
            pointer.next = self.head 
            node1 = None 


            
###############################################################
#
#    Deleting data after an element in circular list
#      --After a particular key element
#       --Go to the node which has to be deleted then
#       --store the address previous to this one in the node next to the one which has to be deleted.
#
#
##############################################################
         
    def Del_after_element(self, pos):     
        key = self.head
        pointer = self.head
        count = 0

        if(pointer != None):
            count += 1
            pointer = pointer.next
        while(pointer != self.head):
            count += 1
            pointer = pointer.next

        if(pos < 1 or pos > count):
            print("\nInavalid pos.")
        
        elif (pos == 1):
            if(self.head.next == self.head):
                self.head = None
            else:
                while(pointer.next != self.head):
                    pointer = pointer.next
                self.head = self.head.next
                pointer.next = self.head
                key = None

        else:
            pointer = self.head
            for i in range(1, pos-1):
                pointer = pointer.next 
            key = pointer.next
            pointer.next = pointer.next.next
            key = None

            
            
#####################################################################
# 
#     Delete at the end
#       --you have to store the address of the node previous to the last node in the 1st node.
#        --Then, free the memory of the node which has to be deleted.
#
#
####################################################################


    def DelEnd(self):
        if(self.head != None):
            if(self.head.next == self.head):
                self.head = None
            else:
                pointer = self.head
                while(pointer.next.next != self.head):
                    pointer = pointer.next
                endnode = pointer.next
                pointer.next = self.head
                endnode = None



# test the code                  
C_ll = CircularLinkedList()

#Insert at Beginning
C_ll.add_begining(72)
C_ll.add_begining(21)
C_ll.add_begining(99)
C_ll.DisplayList()



#Insert at End
C_ll.insert_at_end(11)
C_ll.insert_at_end(22)
C_ll.insert_at_end(32)
C_ll.insert_at_end(45)
C_ll.DisplayList()



#Insert after an element
C_ll.Insert_at_position(519, 2)
C_ll.DisplayList()



#Delete at Beginning
C_ll.Del_at_Begining()
C_ll.DisplayList()  


#Delete after an element
C_ll.Del_after_element(3)
C_ll.DisplayList()


#Delete at end
C_ll.DelEnd()
C_ll.DisplayList() 
        
 
        
```
